```python
#funkcja zwracana przez funkcje

def f(w):
    def g(p):
        return p ** w
    return g

kwadrat = f(2) #kwadrat to funkcja g z w =2, ale skad g pamieta ze w=2? Mechanizm domknięcia
print(kwadrat(5))
```

```python
def uśredniacz():
    '''Zwraca średnią kroczącą.'''
    wartości = []
    
    def średnia(wartość):
        wartości.append(wartość)
        return sum(wartości) / len(wartości)
    
    return średnia

oceny = uśredniacz()
```

```python
oceny(5)
```

```python
oceny(4)
```

```python
oceny(1)
```

```python
oceny.__code__.co_freevars #to pamieta (te zmienną)
```

```python
oceny.__closure__[0].cell_contents #wartosci tej zmiennej
```

<h4> Dekoratory

```python
def funkcja_modyfikujaca(f):
    def wrapper():
        print("Jak tam?")
        f()
    return wrapper

@funkcja_modyfikujaca
def przywitanie():
    print("Dobrze")
    
przywitanie()
```

```python
from functools import lru_cache #implementuje memoizacje (zapamietywanie posrednich wynikow)

@lru_cache()
def fib(n):
    if n == 0 or n == 1:
        return n 
    return fib(n - 1) + fib(n - 2)

print(fib(25))
```

<h4> Generatory

```python
def myGenerator():
    print('Pierwszy element')
    yield 1

    print('Drugi element')
    yield 2

    print('Trzeci element')
    yield 30
```

```python
gen = myGenerator()
```

```python
next(gen) 
```

```python
next(gen) 
```

```python
#funkcja generatora

def fib():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b
        
fibonacci = fib()

[next(fibonacci) for _ in range(15)]
```

```python
a1 = (x*x for x in range(50))
print(a1)
```

```python
a2 = [x*x for x in range(50)]
print(a2)
```
